The value counts of the **fees_status** column show that about 70% of the fees were successfully received and the rest 30% were either cancelled/rejected/confirmed. In order to calculate the revenue generated per user we should only focus on the rows with accepted status. 
If we further check the value counts with the **fees_type** and **fees_status** combined, we see that about 49% of the revenue is generated due to instant payment, 7.4% due to incidents and 14% due to postponing. 

```python
#Filtering the fees that were only received successfully
fees_success_df = fees_subset[fees_subset['fees_status'] == 'accepted']

#Set a flag generated due to incidents
fees_success_df = fees_success_df.copy()
fees_success_df['incident_count'] = np.where(fees_success_df['fees_type'] == 'incident', 1, 0)

#Grouping the accepted fees per cash request id
fees_grp_df = fees_success_df[['cash_request_id',
                               'fees_id',
                               'incident_count',
                               'fees_amount']].groupby(['cash_request_id']).agg({'fees_id':'nunique',
                                                                                 'incident_count':'sum',
                                                                                 'fees_amount':'sum'}).reset_index()

#Rename the column fees_id
fees_grp_df = fees_grp_df.rename(columns={'fees_id':'total_fees_count'})

fees_grp_df
```

# Cohort Analysis

## Merge the datasets together and change the relevant datatypes

```python
merged_df = pd.merge(cash_subset,fees_grp_df,on='cash_request_id',how='left')

#Fill the null rows with zero which shows that some cash requests did not generate a fee for the user
merged_df.fillna({'fees_amount': 0, 'incident_count': 0, 'total_fees_count':0}, inplace=True)
```

```python
#Change the datatypes
merged_df['cr_created_at'] = merged_df['cr_created_at'].apply(lambda x:pd.to_datetime(x))
merged_df['cr_updated_at'] = merged_df['cr_updated_at'].apply(lambda x:pd.to_datetime(x))
merged_df['user_id'] = merged_df['user_id'].astype('int').astype('str')
merged_df['cash_request_id'] = merged_df['cash_request_id'].astype('str')
merged_df['incident_count'] = merged_df['incident_count'].astype('int')
merged_df['total_fees_count'] = merged_df['total_fees_count'].astype('int')
```

## Generate the columns relevant to create the cohort tables

```python
#Create a column storing the first cash request date
merged_df['first_cash_request_date'] = merged_df.groupby('user_id')['cr_created_at'].transform('min')

# 1. Extract year-month from the first cash request date to get the cohort month
merged_df['cohort'] = merged_df['first_cash_request_date'].dt.tz_localize(None).dt.to_period('M')

# 2. Extract year-month of each cash request
merged_df['cr_year_month'] = merged_df['cr_created_at'].dt.tz_localize(None).dt.to_period('M')

# 3. Calculate the cohort index (months since first purchase)
merged_df['cohort_index'] = ((merged_df['cr_year_month'] - merged_df['cohort']).apply(lambda x: x.n))

merged_df
```

# Cohort Table Creation and Visualizations

```python
#Table to show the number of users per cohort
cohort_user_cnt = merged_df[['cohort','user_id']].groupby('cohort').nunique().reset_index()
cohort_user_cnt.rename(columns={'user_id':'num_users'},inplace=True)
cohort_user_cnt
```

```python
# Create the bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='cohort', y='num_users',data=cohort_user_cnt, 
                 hue='cohort', palette='viridis', legend=False)

# Add value labels on top of the bars
for bar in ax.containers:
    ax.bar_label(bar, fmt='%d', label_type='edge', padding=3)

# Customize the plot
plt.title("# Users per Cohort")
plt.xlabel("Cohort")
plt.ylabel("# Users")
plt.xticks(rotation=45) 
plt.show()
```

```python
#Cohort table to show the frequency of cash requests per cohort over time
freq_cohort_table = merged_df.pivot_table(
    index='cohort', 
    columns='cohort_index', 
    values='cash_request_id', 
    aggfunc='count', 
    fill_value = 0

)
```

```python
#Cohort table to show the fees generated per cohort over time
fees_cohort_table = merged_df.pivot_table(
    index='cohort', 
    columns='cohort_index', 
    values='fees_amount', 
    aggfunc='sum', 
    fill_value = 0
)
```

```python
#Check the incident rate per cohort
incidents_df = merged_df[['cohort',
                          'incident_count',
                          'total_fees_count']].groupby('cohort').agg({'incident_count':'sum',
                                                                      'total_fees_count':'sum'}).reset_index()

incidents_df['cohort_incident_rate'] = (incidents_df['incident_count'] / incidents_df['total_fees_count']) * 100

incidents_df['cohort_incident_rate'] = incidents_df['cohort_incident_rate'].round(2)

incidents_df
```

```python
# Visualize the Frequency of Service Usage of cohorts over time
plt.figure(figsize=(10, 6))
sns.heatmap(freq_cohort_table, annot=True, fmt="d", cmap="YlGnBu", cbar_kws={'label': 'Frequency of Service Usage'})
plt.title("Cohort Analysis - Frequency of Cash Request")
plt.xlabel("Months Since Cohort")
plt.ylabel("Cohort Month")
plt.show()
```

```python
# Visualize the Revenue per cohort over time
plt.figure(figsize=(10, 6))
sns.heatmap(fees_cohort_table, annot=True, fmt=".0f", cmap="YlGnBu", cbar_kws={'label': 'Revenue (USD)'})
plt.title("Cohort Analysis - Revenue")
plt.xlabel("Months Since Cohort")
plt.ylabel("Cohort Month")
plt.show()
```

```python
# Visualize the incident rate per cohort
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='cohort', y='cohort_incident_rate',data=incidents_df, hue='cohort', palette='viridis', legend=False)

# Add value labels on top of the bars
for bar in ax.containers:
    ax.bar_label(bar, fmt='%.1f', label_type='edge', padding=3)

# Customize the plot
plt.title("Incident Rate per Cohort")
plt.xlabel("Cohort")
plt.ylabel("Incident Rate")
plt.xticks(rotation=45) 
plt.show()
```
