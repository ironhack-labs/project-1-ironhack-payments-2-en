```python
#Creating a column with month and year of each request created and counting the frequency using groupby over cohort month & request month
cash['request_month'] = cash['created_at'].dt.to_period('M')
usage_frequency = cash.groupby(['cohort_month', 'request_month']).size().reset_index(name='usage_count')

#display(usage_frequency)

usage_frequency.head(50)
```

```python
#Count the number of users in each cohort
user_count_cohort = cash.groupby("cohort_month")["user_id"].nunique().reset_index(name="user_count")
print(user_count_cohort)
```

```python
#visualize Frquency of Service Usage (Metric 1)
import seaborn as sns

# Step 1: Pivot the data for plotting
pivot_usage_frequency = usage_frequency.pivot(index='request_month', columns='cohort_month', values='usage_count')

# Step 2: Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_usage_frequency, cmap='YlGnBu', annot=True, fmt='g')
plt.title('Frequency of Service Usage by Cohort Over Time (Heatmap)')
plt.ylabel('Request Month')
plt.xlabel('Cohort Month')
plt.tight_layout()
plt.show()
```

Task 2: Incident Rate

Determine the incident rate, specifically focusing on payment incidents, for each cohort. Identify if there are variations in incident rates among different cohorts.

```python
#Turning the fees cashrequest id to int and renaming the id from cash df to match the one from fees.
#Merging the fees and cash dfs to create a df with cash requests that had fees
fees["cash_request_id"] = fees["cash_request_id"].astype(int)
cash.rename(columns={'id': 'cash_request_id'}, inplace=True)
cash_fees = pd.merge(cash, fees[["cash_request_id","type",]], left_on="cash_request_id", right_on="cash_request_id", how="inner")

display(cash_fees)
```

```python
# Step 1: Calculate the total number of requests per cohort
cohort_data = cash.groupby('cohort_month').agg(
    total_requests=('user_id', 'size')  # Total number of requests (rows) per cohort
).reset_index()

display(cohort_data)
```

```python
# Step 2: Calculate the number of incidents (where 'type' is 'incident') per cohort
incident_data = cash_fees[cash_fees['type'] == 'incident'].groupby('cohort_month').agg(
    total_incidents=('user_id', 'size')  # Count of incidents per cohort
).reset_index()
display(incident_data)
```

```python
# Step 3: Merge the two datasets
cohort_data = pd.merge(cohort_data, incident_data, on='cohort_month', how='left')
```

```python
# Step 4: Fill NaN values in 'total_incidents' with 0 (for cohorts with no incidents)
cohort_data['total_incidents'] = cohort_data['total_incidents'].fillna(0)
```

```python
# Step 5: Calculate the incident rate (incidents / total requests)
cohort_data['incident_rate'] = cohort_data['total_incidents'] / cohort_data['total_requests']

# Display the results
print(cohort_data)
```

```python

# Step 6: Plot the incident rate for each cohort
cohort_data.plot(kind='bar', x='cohort_month', y='incident_rate', figsize=(10,6), color='skyblue')
plt.title('Incident Rate by Cohort')
plt.ylabel('Incident Rate')
plt.xlabel('Cohort Month')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

3. Revenue Generated by the Cohort: Calculate the total revenue generated by each cohort over months to assess the financial impact of user behavior.

```python
#Renaming the status column in fees to not have two columns with the same name after the merge
fees = fees.rename(columns={"status": "status_fees"})
```

```python
fees.columns
```

```python
#changing the date type of the "paid_at" column which specifies when a fee has been paid back
fees['paid_at'] = pd.to_datetime(fees['paid_at'], errors="coerce")
fees['paid_month'] = fees['paid_at'].dt.to_period('M')
```

```python
#merge the data sets with the necessary columns
cash_fees = pd.merge(cash, fees[['cash_request_id', 'total_amount', 'status_fees', 'type',"paid_month"]], on='cash_request_id', how='left')
display(cash_fees)
```

```python
#Filter for accepted fees (revenue-generating)
revenue_data = cash_fees[cash_fees['status_fees'] == 'accepted']

# Display to check the filtered data
display(revenue_data)
```

```python
revenue_data["total_amount"].unique()
```

```python
# Group by cohort_month and request_month to calculate the total revenue for each cohort over time
revenue_by_cohort = revenue_data.groupby(['cohort_month', 'paid_month']).agg(
    total_revenue=('total_amount', 'sum')  # Sum of all revenue (fees) for each cohort
).reset_index()

# Display the final result
display(revenue_by_cohort)

total_revenue = revenue_by_cohort["total_revenue"].sum()
print(total_revenue)
```

```python
# Pivot the data for the heatmap (based on 'cohort_month' and 'request_month')
pivot_revenue_by_cohort = revenue_by_cohort.pivot(index="paid_month", columns='cohort_month', values='total_revenue')

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_revenue_by_cohort, cmap='YlGnBu', annot=True, fmt='.2f')  # Format the numbers to 2 decimal places
plt.title('Total Revenue by Cohort Over Time (Heatmap)')
plt.ylabel("Month of the fee's payment")
plt.xlabel('Cohort Month')
plt.tight_layout()
plt.show()
```

```python
#4. Probability that a user will utilize the service after he has used it for the first time by cohorts


#Create a flag for subsequent usage (True if user has more than 1 request after their first)
cash['subsequent_usage'] = cash['request_month'] > cash['cohort_month']

# Group by user and check if they have any subsequent usage (i.e., more than one request)
subsequent_usage_flag = cash.groupby('user_id')['subsequent_usage'].max().reset_index()
subsequent_usage_flag.columns = ['user_id', 'used_again']

# Merge this subsequent usage flag back into the first usage data
cash = pd.merge(cash, subsequent_usage_flag, on='user_id', how='left')
```

```python
#Calculate the number of users per cohort and those who used the service again
cohort_analysis = cash.groupby('cohort_month').agg(
    total_users=('user_id', 'nunique'),        # Total number of unique users in the cohort
    users_used_again=('used_again', 'sum')     # Number of users who used the service again
).reset_index()

# Calculate the probability that a user will use the service again
cohort_analysis['probability_used_again [in %]'] = cohort_analysis['users_used_again'] / cohort_analysis['total_users']*100

# Display the results
print(cohort_analysis[['cohort_month', 'total_users', 'users_used_again', 'probability_used_again [in %]']])
```
