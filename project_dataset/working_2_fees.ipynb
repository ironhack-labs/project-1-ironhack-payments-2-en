```python
# Dropped the columns that I didn't need
df1 = df1.drop(columns=["reason","created_at","updated_at","paid_at","from_date","to_date","charge_moment"])
df1
```

```python
# for each category how many money company turned back based on the total amount.
 fee_payments = df1.groupby("status")["total_amount"].sum()
fee_payments
```

```python
import matplotlib.pyplot as plt

status = ["accepted", "cancelled", "confirmed", "rejected"]
amounts = [74210.0, 24690.0, 440.0, 5970.0]  # Corresponding total amounts

fig, ax = plt.subplots(figsize=(9, 6))  

# Plot bar chart
ax.bar(status, amounts, color='skyblue', edgecolor="black")

ax.set_title('Categorical total fees', fontsize=16)
ax.set_xlabel('Status', fontsize=12)
ax.set_ylabel('Total Amount', fontsize=12)


plt.tight_layout()  
plt.show()
```
